```python
def twintConfig(search_string):
    c = twint.Config()
    c.Search = search_string[1]
    c.Limit=500
    c.Lang = 'en'
    c.Pandas = True
    c.Pandas_au = True          
    c.Pandas_clean=True
    c.Since='2016-01-01'
    c.Until='2022-12-31'
    twint.run.Search(c)
```

```python
def Run_Twint(search_vals):
    
    #set empty dataframe for join
    out_df= pd.DataFrame()
    
    for zarr in search_vals.items():
        print ("running for search item: "+zarr[0]+"\n")
        print ("Search string: "+zarr[1]+"\n")
        
        #run twint
        twintConfig(zarr)
        
        #get dataframe
        tweets_df = twint.storage.panda.Tweets_df
        
        #join Dataframes and create 'Candi' column
        tweets_df["Zarr_tweet"]= zarr[0]
        out_df = pd.concat([out_df,tweets_df])
        
    return out_df
```

```python
twint_df = Run_Twint(name)
```
